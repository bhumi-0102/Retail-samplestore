```python
#Lets see how sub-categories are distributed wrt to category
plt.figure(figsize=(16,8))
plt.bar('Sub-Category','Category',data=df,color='g')
plt.show()
```

From this graph, one can easily makeout which Category & Sub-Category to choose when they are looking to purchase a product

```python
plt.figure(figsize=(12,10))
df['Sub-Category'].value_counts().plot.pie(autopct="%1.1f%%")
plt.show()
```

The store has wide variety of Office Supplies especially in Binders and Paper department.

```python
df.groupby('Sub-Category')['Profit','Sales'].agg(['sum']).plot.bar()
plt.title('Total Profit and Sales per Sub-Category')
```

Highest profit is earned in Copiers while Selling price for Chairs and Phones is extremely high compared to other products.

Another interesting fact- people dont prefer to buy Tables and Bookcases from Superstore. Hence these departments are in loss.

```python
plt.figure(figsize=(10,8))
df['Region'].value_counts().plot.pie(autopct = '%1.1f%%')
plt.show()
```

```python
#Region wise sales plot
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(df['Region'],df['Sales'], color = 'r')
plt.title("Region VS Sales")
plt.xlabel("Regions")
plt.ylabel("Sales")
plt.show()
```

Observation:- The South region is leading the sales followed by the East, West and Central.

### Category wise sales in Each region

```python
plt.figure(figsize=[12,8])
ax = sns.barplot(x="Region", y="Sales", hue="Category", data=df, palette="Paired")
```

Observation:- In each and every region, sales for 'Office Supplies' are very poor. Furniture and Technology are well ahead.

### Segment wise sales in Each region

```python
plt.figure(figsize=[12,8])
ax = sns.barplot(x="Region", y="Sales", hue="Segment", data=df, palette="mako")
```

#### Observation:- In case of sales not much difference is seen based on the Segments for any region. Overall 'Corporate' is leading a bit.

## Sales vs Subcategory

```python
plt.figure(figsize=[20,10])
plt.bar("Sub-Category","Sales",data=df, color = 'y')
plt.title("Sub_Category VS Sales",fontsize=18)
plt.xlabel("Subcategory", fontsize=16)
plt.ylabel("Sales",fontsize=16)
plt.show()
```

#### Observation:- As we can see, the 'Machines' sub-category produces maximum sales followed by the 'Copires' sub-category. 'Fasteners' are lagging behind of all.
